```python
def group_files(experiment,dry_run=True):
    ''' Copies files into a new provided directory tree structure '''
    metadata = experiment.metadata
    parent = experiment.experiment_directory
    total = len(metadata)
    for i,row in metadata.iterrows():
        filename = row["filename"]

        old = f"{parent}/Images/{filename}"

        cell = row["cell"]
        coating = row["coating"]
        well = row["well"]
        field = row["field"]
        ch = row["ch"]
        grouped = f"{parent}/grouped"

        # TODO: make this hierarchy tweakable
        new = f"{grouped}/{cell}/{coating}/{well}/f{field}/ch{ch}/{filename}"

        if not dry_run:
            clear_output() # We need to find a way to incorporate this inside the function
            get_loading_bar(i,total)
            os.makedirs(os.path.dirname(new), exist_ok=True)
            shutil.copy(old, new)
        else:
            print(old)
            print(new)
```

```python
from IPython.display import clear_output
def get_loading_bar(i,length):
    if i > length:
        i = i+1
        percentage = round(i/length*100)
        bar_length = 10

        current_bar = round(i/length * bar_length)
        remaining_bar = bar_length - current_bar

        bar_progress = "* " * current_bar
        bar_gap = "_ " * remaining_bar
        bar = f"[{bar_progress}{bar_gap}] {percentage}% [{i}/{length}]"
        print(bar)
```

# Creating a language to describe rules to generate experiment info
- rule = \<location> \<condition_type> \<condition>  
- location_type = \[well,col,row] 
- condition_type = \[cell,media,coating]

Location type is inferred by location value!


```python
import re
def parse_rule(rule):
    condition_types = ["cell", "media","coating"]

    tokens = rule.split(" ")
    location,condition_type,*condition = tokens
    condition = " ".join(condition)
    
    location_type = parse_location(location)
    if not location_type:
        rule = False
    elif condition_type not in condition_types:
        rule =  False
    else:
        # Probably a better place to put this...?
        if location_type == "col":
            location = int(location)
        rule = {
            "location_type" : location_type,
            "location": location,
            "condition_type": condition_type,
            "condition" : condition
        }
    return rule
        
def parse_location(location):
    location_type = ""
    well_re = "^[a-zA-Z]{1}\d+$"
    row_re = "^[a-zA-Z]{1}$"
    col_re = "^\d*$"
    
    well = re.search(well_re,location)
    row = re.search(row_re,location)
    col = re.search(col_re,location)
    
    if well:
        location_type = "well"
    elif row:
        location_type = "row_letter"
    elif col:
        location_type = "col"
    else:
        location_type = False
    
    return location_type

def apply_rule(rule,exp):
    metadata = exp.metadata
    rule = parse_rule(rule)
    if rule:
        loc_type,loc,con_type,con = rule.values()
        metadata.loc[metadata[loc_type] == loc,con_type] = con
    else:
        print("Improper Rule")
    return metadata
    
```
